```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from IPython.core.display import display, HTML
```

```python
response = requests.get(
    "https://portal.trf1.jus.br/data/files/8E/F2/DD/09/D637371031ADF637E52809C2/precat_rios%20federais%202021%20excel%20-%20Resolu__o%20303%20do%20CNJ.htm"
)

soup = BeautifulSoup(response.text, "html.parser")
```

```python
# HTML(str(soup))
trs = soup.select("table tr")

trs = [
    {
        "PRECATORIO": tr.select_one("td:nth-child(2)").text,
        "VALOR": tr.select_one("td:nth-child(3)").text,
    }
    for tr in trs
    if len(tr.select("td")) == 4
]
trs
```

```python
df = pd.DataFrame(trs, columns=["PRECATORIO", "VALOR"])

df["PRECATORIO"].replace("", np.nan, inplace=True)
df["VALOR"].replace("", np.nan, inplace=True)

df["PRECATORIO"].replace(["[\sA-Za-z]+"], [np.nan], regex=True, inplace=True)

df.dropna(subset=["PRECATORIO", "VALOR"], inplace=True)

df
```

```python
df.sort_values(by=["VALOR"], ascending=True, inplace=True)
df.to_excel("/home/jonatas/Downloads/list-values-2021.xlsx")
```
