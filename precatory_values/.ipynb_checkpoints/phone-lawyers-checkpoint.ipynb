```python
import pandas as pd

l1 = pd.read_excel('/home/jonatas/Downloads/finding-lawyers-phones.xlsx', usecols="A:B")
l1
```

```python
l1.columns
```

```python
import numpy as np

l1.TELEFONE = l1.TELEFONE[np.logical_or(l1.TELEFONE.str.len() == 14 , l1.TELEFONE.str.len() == 12)]
l1 = l1[l1['TELEFONE'].notna()]
l1
```

```python
l2 = pd.read_excel("list-values.xlsx", usecols="A:I")
del l2['0_x']
del l2['Unnamed: 0']
l2['2_x'].astype(str)
l2
```

```python
# print(l2['3_x'].isin(l1['ADVOGADO']))
l2 = l2[l2['3_x'].notna()]

# l2['3_x'][l2['3_x'].str.contains("foo", na=False)]
# l1['ADVOGADO'].isin(l2['3_x'])

l1.columns = ['3_x', 'TELEFONE']
fl = pd.merge(l1, l2, on='3_x', how='outer')
fl = fl[fl['TELEFONE'].notna()]
fl = fl[fl['2_y'].notna()]
fl = fl.sort_values('2_y', ascending=False)
fl.columns = ['ADVOGADO', 'TELEFONE', 'PRECATÓRIOS', 'REQUERENTE', 'REQUERIDO', 'DEPRECANTE', 'VALORES', 'TIPO']

fl['PRECATÓRIOS'].astype(str)

fl['PRECATÓRIOS'] = fl['PRECATÓRIOS'].apply(lambda x: '{:.0f}'.format(x))

fl = fl.drop_duplicates(keep=False, inplace=True)

fl
```

```python
fl.to_excel("list-values-with-phone.xlsx")
```
